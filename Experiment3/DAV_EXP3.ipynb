 **Theory:**

Regression is a statistical technique used for modeling the relationship between a dependent variable (or response variable) and one or more independent variables (or predictor variables). The goal of regression analysis is to understand how changes in the independent variables are associated with changes in the dependent variable. It is widely used in various fields for prediction, forecasting, and understanding the relationships between variables.


**Multiple Linear Regression:**

Multiple Linear Regression is a statistical method used to model the relationship between a dependent variable and two or more independent variables. The model is an extension of simple linear regression, allowing for a more complex understanding of how multiple factors influence the variation in the dependent variable.

Multiple Independent Variables: Involves two or more independent variables, allowing for a more comprehensive analysis of the relationship.

Fitting the Model: The model is fitted by minimizing the sum of squared differences between predicted and observed values (least squares estimation).

Assumptions: Similar assumptions to simple linear regression, including linearity, independence of errors, homoscedasticity, and normality of errors.

Evaluation Metrics: Common evaluation metrics include R-squared, Mean Squared Error (MSE), Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE).

Why Study Multiple Linear Regression:

1. Real-world Complexity: Many real-world scenarios involve multiple factors influencing an outcome. MLR allows us to model such complex relationships.

2. Improved Predictions: MLR can provide more accurate predictions compared to simple linear regression when multiple variables contribute to the outcome.

3. Understanding Variable Interactions: Helps in understanding how different independent variables interact with each other in influencing the dependent variable.

4. Variable Importance: Identifies which independent variables are most important in explaining the variability in the dependent variable.

5. Model Interpretability: Allows for a more nuanced interpretation of the relationships between variables.

### Multiple Linear Regression in Python

```r
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```r
data = pd.read_csv('/content/advertising.csv')
```

```r
data.head()
```

```r
data.describe()
```

```r
x = data['TV']
y = data['Sales']
plt.scatter(x, y)
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```r
x = data['Radio']
y = data['Sales']
plt.scatter(x, y)
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()
```

```r
x = data['Newspaper']
y = data['Sales']
plt.scatter(x, y)
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.show()
```

```r
x = data[['TV', 'Radio', 'Newspaper']]
y = data['Sales']
```

```r
from sklearn.model_selection import train_test_split
x_train ,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```r
from sklearn.linear_model import LinearRegression
```

```r
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```r
print("Intercept: ", mlr.intercept_)
print("Coefficients:")
list(zip(x, mlr.coef_))
```

```r
y_pred_mlr= mlr.predict(x_test)

print("Prediction for test set: {}".format(y_pred_mlr))
```

```r
mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff.head()
```

```r
from sklearn import metrics
meanAbErr = metrics.mean_absolute_error(y_test, y_pred_mlr)
meanSqErr = metrics.mean_squared_error(y_test, y_pred_mlr)
rootMeanSqErr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_mlr))
print('R squared: {:.2f}'.format(mlr.score(x,y)*100))
print('Mean Absolute Error:', meanAbErr)
print('Mean Square Error:', meanSqErr)
print('Root Mean Square Error:', rootMeanSqErr)
```

```r
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = data['TV']
y = data['Newspaper']
z = data['Radio']

ax.scatter(x, y, z)

A = np.vstack([x, y, np.ones_like(x)]).T
plane_coef, _, _, _ = np.linalg.lstsq(A, z, rcond=None)

x_plane, y_plane = np.meshgrid(x, y)
z_plane = plane_coef[0] * x_plane + plane_coef[1] * y_plane + plane_coef[2]

ax.plot_surface(x_plane, y_plane, z_plane, alpha=0.5)

ax.set_xlabel('TV')
ax.set_ylabel('Newspaper')
ax.set_zlabel('Radio')
plt.title('Multiple Linear Regression')

plt.show()
```

```r
data <- read.csv('/content/advertising.csv')

head(data)

dim(data)
```

```r
x <- data$TV
y <- data$Sales

plot(x, y, main="Scatter Plot of TV vs. Sales", xlab="TV", ylab="Sales", col="green")
```

```r
x <- data$Radio
y <- data$Sales

plot(x, y, main="Scatter Plot of Radio vs. Sales", xlab="Radio", ylab="Sales", col="green")
```

```r
x <- data$Newspaper
y <- data$Sales

plot(x, y, main="Scatter Plot of Newspaper vs. Sales", xlab="Newspaper", ylab="Sales", col="green")
```

```r
library(dplyr)

selected_data <- data %>% select(TV, Radio, Newspaper, Sales)

set.seed(100)
train_data <- selected_data %>% sample_frac(0.7)
test_data <- anti_join(selected_data, train_data)
```

```r
mlr_model <- lm(Sales ~ TV + Radio + Newspaper, data = train_data)

summary(mlr_model)
```

```r
predictions <- predict(mlr_model, newdata = test_data)

print(predictions)
```

```r
mlr_diff <- data.frame('Actual value' = test_data$Sales, 'Predicted value' = predictions)

head(mlr_diff)
```

```r
meanAbErr <- mean(abs(test_data$Sales - predictions))
meanSqErr <- mean((test_data$Sales - predictions)^2)
rootMeanSqErr <- sqrt(meanSqErr)

r_squared <- 1 - (sum((test_data$Sales - predictions)^2) / sum((test_data$Sales - mean(test_data$Sales))^2))

cat('R squared:', format(r_squared * 100, digits = 2), '\n')
cat('Mean Absolute Error:', meanAbErr, '\n')
cat('Mean Square Error:', meanSqErr, '\n')
cat('Root Mean Square Error:', rootMeanSqErr, '\n')
```
